```python
#Use the data that is in the data folder to print a candlestick chart
import plotly.graph_objects as go
import pandas as pd
df = pd.read_csv("data/30minutes_btc_01_01_19_01_01_22.csv")
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
fig.write_html('candlestick.html', auto_open=True)
```

```python
import pandas as pd
df = pd.read_csv("data/30minutes_btc_01_01_21_20_12_22.csv")
df = df[df['Date'] > '2021-12-27 10:00:00']
df = df[df['Date'] < '2021-12-29 14:00:00']
import plotly.graph_objects as go
from plotly.subplots import make_subplots
fig = go.Figure()
fig = make_subplots(rows=1, cols=1)

# candlestick
fig.append_trace(
    go.Candlestick(
        x=df.Date,
        open=df.Open,
        high=df.High,
        low=df.Low,
        close=df.Close,  
    ), row=1, col=1
)
fig.append_trace(
    go.Scatter(
        x=df.Date,
        y=df.kijun_sen,
        mode='lines',
        name='kijun_sen',
    ), row=1, col=1
)
fig.append_trace(
    go.Scatter(
        x=df.Date,
        y=df.tenkan_sen,
        mode='lines',
        name='tenkan_sen',
    ), row=1, col=1
)
fig.append_trace(
    go.Scatter(
        x=df.Date,
        y=df.MA100,
        mode='lines',
        name='MA100',
    ), row=1, col=1
)
fig.update(layout_xaxis_rangeslider_visible=False)
fig.show()
#save the figure to png
fig.write_image("candlestick_2021_12_29_10_14.png", width=1920, height=1080, scale=1)
```
