```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("data/30minutes_btc_01_01_21_20_12_22.csv")
```

```python
current = 0
last_buy = 0
last_sell = 0
last_nothing = 0
for i in range(50,len(df)):
    if(i % 1000 == 0):
        print(i)
    df = pd.read_csv("data/30minutes_btc_01_01_21_20_12_22.csv")
    if(df["final"][i] > df["Close"][i]*1.02):
        #choose only the last 30 candles
        if (current == 1 and i - last_buy < 3):
            continue
        else:
            current = 1
            last_buy = i
            df = df.iloc[i-50:i]
            #reset the index
            df = df.reset_index(drop=True)
            #create the figure
            fig = go.Figure()
            fig = make_subplots(rows=1, cols=1)

            # candlestick
            fig.append_trace(
                go.Candlestick(
                    x=df.Date,
                    open=df.Open,
                    high=df.High,
                    low=df.Low,
                    close=df.Close,  
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.kijun_sen,
                    mode='lines',
                    name='kijun_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.tenkan_sen,
                    mode='lines',
                    name='tenkan_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.MA100,
                    mode='lines',
                    name='MA100',
                ), row=1, col=1
            )
            fig.update(layout_xaxis_rangeslider_visible=False)
            #save the figure to png
            fig.write_image("process/buy/"+str(i)+".png", width=1920, height=1080, scale=1) 

    elif(df["final"][i] < df["Close"][i]*0.98):
        #choose only the last 30 candles
        if (current == -1 and i - last_sell < 3):
            continue
        else:
            current = -1
            last_sell = i
            df = df.iloc[i-50:i]
            #reset the index
            df = df.reset_index(drop=True)
            fig = go.Figure()
            fig = make_subplots(rows=1, cols=1)

            # candlestick
            fig.append_trace(
                go.Candlestick(
                    x=df.Date,
                    open=df.Open,
                    high=df.High,
                    low=df.Low,
                    close=df.Close,  
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.kijun_sen,
                    mode='lines',
                    name='kijun_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.tenkan_sen,
                    mode='lines',
                    name='tenkan_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.MA100,
                    mode='lines',
                    name='MA100',
                ), row=1, col=1
            )
            fig.update(layout_xaxis_rangeslider_visible=False)
            #save the figure to png
            fig.write_image("process/sell/"+str(i)+".png", width=1920, height=1080, scale=1) 
    else:
        if (current == 0 and i - last_nothing < 20):
            continue
        else:
            current = 0
            last_nothing = i
            #choose only the last 30 candles
            df = df.iloc[i-50:i]
            #reset the index
            df = df.reset_index(drop=True)
            fig = go.Figure()
            fig = make_subplots(rows=1, cols=1)

            # candlestick
            fig.append_trace(
                go.Candlestick(
                    x=df.Date,
                    open=df.Open,
                    high=df.High,
                    low=df.Low,
                    close=df.Close,  
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.kijun_sen,
                    mode='lines',
                    name='kijun_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.tenkan_sen,
                    mode='lines',
                    name='tenkan_sen',
                ), row=1, col=1
            )
            fig.append_trace(
                go.Scatter(
                    x=df.Date,
                    y=df.MA100,
                    mode='lines',
                    name='MA100',
                ), row=1, col=1
            )
            fig.update(layout_xaxis_rangeslider_visible=False)
            #save the figure to png
            fig.write_image("process/nothing/"+str(i)+".png", width=1920, height=1080, scale=1) 
```
